```python
def outliers_detection(df): #Creating the function for detecting the outliers in the dataset
    outliers = pd.DataFrame() #Creating the dataframe to append the outliers
    for i in df.describe().columns:
        q1 = df[i].quantile(0.25) #Finding the first quantile of column
        q3 = df[i].quantile(0.75) #Finding the third quantile of column

        IQR = q3 - q1 #Finding the Interquartile range

        lower_bound = q1 - (1.5 * IQR) #finding the lower bound of column
        upper_bound = q3 + (1.5 * IQR) #finding the upper bound of column

        out = df[(df[i]<lower_bound) | (df[i]>upper_bound)] #Finding the outliers in the dataset
        outliers = pd.concat([outliers,out]) #Append the outliers to dataframe

    return outliers #Return the outliers from the dataset
```

```python
outliers = outliers_detection(df)
pca_outlier = outliers.drop_duplicates().reset_index().drop('index',axis=1) #Create a separate dataframe for outliers by droping duplicates
pca_outlier
```

**Dimensionality Reduction**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# standardizing the features
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
#implementing pca
pca = PCA()
pca_data = pca.fit_transform(scaled_df)
```

```python
explained_variance = pca.explained_variance_ratio_
plt.plot(explained_variance)
plt.xlabel("Number of Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_df) # transforming the data again
print(reduced_data)
```

**Clustering with Data**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
cluster_labels_value = kmeans.labels_
```

```python
# using scatterplot to visualize scatter plot
plt.scatter(df["Type"], df["Alcohol"], c=cluster_labels_value)  # Replace features with your desired axes
plt.xlabel("Type")
plt.ylabel("Alcohol")
plt.title("K-means Clustering Results")
plt.show()
```

```python
silhouette_coeff_value = silhouette_score(df, cluster_labels_value)
print("Silhouette Score:", silhouette_coeff_value)
```

```python
db_score = davies_bouldin_score(df, cluster_labels_value)
print("Davies-Bouldin Score:", db_score)
```
